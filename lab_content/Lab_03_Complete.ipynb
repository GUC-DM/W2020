## Introduction to Pandas

**pandas** is a Python package providing fast, flexible, and expressive data structures designed to work with *relational* or *labeled* data both. It is a fundamental high-level building block for doing practical, real world data analysis in Python. 

pandas is well suited for:

- Tabular data with heterogeneously-typed columns, as in an SQL table or Excel spreadsheet
- Ordered and unordered (not necessarily fixed-frequency) time series data.
- Arbitrary matrix data (homogeneously typed or heterogeneous) with row and column labels
- Any other form of observational / statistical data sets. The data actually need not be labeled at all to be placed into a pandas data structure


Key features:
    
- Easy handling of **missing data**
- **Size mutability**: columns can be inserted and deleted from DataFrame and higher dimensional objects
- Automatic and explicit **data alignment**: objects can be explicitly aligned to a set of labels, or the data can be aligned automatically
- Powerful, flexible **group by functionality** to perform split-apply-combine operations on data sets
- Intelligent label-based **slicing, fancy indexing, and subsetting** of large data sets
- Intuitive **merging and joining** data sets
- Flexible **reshaping and pivoting** of data sets
- **Hierarchical labeling** of axes
- Robust **IO tools** for loading data from flat files, Excel files, databases, and HDF5
- **Time series functionality**: date range generation and frequency conversion, moving window statistics, moving window linear regressions, date shifting and lagging, etc.

## Pandas Data Structures

### Series

A **Series** is a one-dimensional array with axis labels collectively known as an index.

Creating a series from an python list. Note how the indicies are automatically generated.

```python
population = pd.Series([1394, 1326, 332, 267, 234, 214, 212, 163, 142, 129])
population
```

We can set the index (axis labels) at creation by passing it to the `index` parameter.

```python
population = pd.Series([1394, 1326, 332, 267, 234, 214, 212, 163, 142, 129],
                       index=['China', 'India', 'United States','Indonesia', 'Pakistan', 'Nigeria', 'Brazil', 'Bangladesh', 'Russia', 'Japan'])

population
```

We can name the series and the index for additional clarity

```python
population.name = 'population'
population.index.name = 'country'
population
```

We can access data by its positional (integer) index, like normal lists, by indexing with the `iloc` attribute

```python
# Indexing
population.iloc[2]
```

```python
# Slicing is also possible
population.iloc[:4]
```

We can access the data associated with a particular index (i.e. axis label), by indexing the `loc` attribute

```python
population.loc['Japan']
```

Similarly, you can also slice using the labels (notice how both the start and stop labels are included when slicing with `loc`)

```python
population.loc['Indonesia' : 'Russia']
```

You can use boolean comparison with a series.

```python
population < 250
```

Which can be used to filter data quite easily. This is called boolean masking.

```python
population[population > 250]
```

### DataFrame

Inevitably, we want to be able to store, view and manipulate data that is *multivariate*, where for every index there are multiple fields or columns of data, often of varying data type.

A `DataFrame` is a tabular data structure, encapsulating multiple series like columns in a spreadsheet. Data are stored internally as a 2-dimensional object, but the `DataFrame` allows us to represent and manipulate higher-dimensional data.

Country, Population (in millions), GDP (nominal, in millions of US dollars) as per the World Bank (2019)

```python
world_stats = {
    'country': ['China', 'India', 'United States','Indonesia', 'Pakistan', 'Nigeria', 'Brazil', 'Bangladesh', 'Russia', 'Japan'],
    'population': [1394, 1326, 332, 267, 234, 214, 212, 163, 142, 129],
    'gdp': [14342903, 2875142, 21427700, 1119191, 278222, 448120, 1839758, 302571, 1699877, 5081770]
}

df_example = pd.DataFrame(world_stats)
df_example
```

Accessing a particular column can be done just by indexing the dataframe 

```python
df_example['country']
```

Note how this returns a series

```python
type(df_example['country'])
```

We can also select multiple columns, by indexing with a list of column names

```python
df_example[['country', 'gdp']]
```

Notice how it returns a dataframe since we still have multiple columns (i.e. two-dimensional data)

```python
type(df_example[['country', 'gdp']])
```

Adding a new column is also straightforward. Here, we're setting a single value for the entire column, but you can assign a python list or a `pd.Series` object to the column as well.

```python
df_example['year'] = 2019
df_example
```

Generate descriptive statistics of numerical columns.

```python
world_df.describe()
```

The population data is a bit hard to read. Let's convert it to be in millions of units.

```python
world_df['population'] = world_df['population'] / 1e6
world_df
```

```python
world_df.describe()
```

We can include non-numeric columns in `describe`'s analysis by specifying `include='all'`

This calculates the number of unique values, the most common value and its frequency for each non-numeric column

```python
world_df.describe(include='all')
```

Let's say we want to know the top 10 richest countries (highest GDP per capita). We can use the `sort_values` function to sort by a particular column.

```python
world_df.sort_values('gdp', ascending=False).head(10)
```

We can filter to specific columns by indexing with a list of column names we'd like to have

```python
sorted_gdp = world_df.sort_values('gdp', ascending=False)
sorted_gdp[['country', 'gdp']].head(10)
```

Let's have a closer look at the region column. We can look at the values under region and their frequency by calling `value_counts()` on the region column.

```python
world_df['region'].value_counts()
```

If we want to filter the dataframe to just the countries in the Northern Africa, we can use index with a boolean mask like before

```python
world_df[world_df['region'] == 'NORTHERN AFRICA']
```

Can we apply statistical methods (mean, median, standard deviation, etc.) on the filtered dataframe?

Yes! It's still a `pd.DataFrame` object.

```python
# try out median(), std() and describe()
world_df[world_df['region'] == 'NORTHERN AFRICA'].mean()
```

Taking it a step further, let's apply statistical analysis per-region instead of the entire world as before. We can use `groupby` and pass it the column we want to group by, and then apply an aggregate function like `mean`.

```python
world_df.groupby('region').mean()
```

We can also use the `agg` (aggregate) function to apply one or more functions to each group

```python
world_df.groupby('region').agg(['mean', 'std'])
```
